```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
basics_url="https://datasets.imdbws.com/title.basics.tsv.gz"
akas_url="https://datasets.imdbws.com/title.akas.tsv.gz"
rating_url="https://datasets.imdbws.com/title.ratings.tsv.gz"
```

```python
basics = pd.read_csv(basics_url, sep='\t', low_memory=False)
```

```python
akas = pd.read_csv(akas_url, sep='\t', low_memory=False)
```

```python
rating = pd.read_csv(rating_url, sep='\t', low_memory=False)
```

```python
basics.replace({'\\N':np.nan})
```

```python
basics.dropna(subset=['runtimeMinutes','genres'],inplace=True)
```

```python
movie_type = basics['titleType'].str.contains('movie', case = False)
```

```python
basics = basics[movie_type]
```

```python
start_date = '2000'
end_date = '2021'
start_year = (basics['startYear'] > start_date) & (basics['startYear'] <= end_date)
```

```python
basics = basics[start_year]
```

```python
is_documentary = basics['genres'].str.contains('documentary',case=False)
basics = basics[~is_documentary]
```

```python
akas.replace({'\\N':np.nan})
```

```python
akas.dropna(subset=['region'],inplace=True)
```

```python
rating.replace({'\\N':np.nan})
```

```python
keepers =basics['tconst'].isin(akas['titleId'])
keepers
```

```python
basics = basics[keepers]
basics
```

```python
us_movie = akas['region'].str.contains('US', case = False)
```

```python
keepers2 =akas['titleId'].isin(rating['tconst'])
keepers2
```

```python
akas = akas[keepers2]
akas
```

```python
basics.to_csv("Data/title_basics.csv.gz",compression='gzip',index=False)
akas.to_csv("Data/title_akas.csv.gz",compression='gzip',index=False)
rating.to_csv("Data/title_rating.csv.gz",compression='gzip',index=False)
```
